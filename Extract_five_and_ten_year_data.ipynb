```python
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt


pd.options.display.float_format = '{:.2f}'.format
```

```python
df_cleaned_2000 = pd.read_csv('./df_cleaned_2000.csv')
# df_cleaned_2000
```

```python
#del df_cleaned_2000['Unnamed: 0']
```

```python
df_cleaned_2000=df_cleaned_2000.drop(columns=['level_0'])
df_cleaned_2000.head()
```

```python
# 5 year dataset
```

```python
df_tech_five_year = df_cleaned_2000.loc[(df_cleaned_2000['date'] >= '2013') & (df_cleaned_2000['sector'] == 'TECHNOLOGY')]
```

```python
opening_day_5year = df_tech_five_year['ticker'].loc[df_tech_five_year['date'] == '2013-01-02'].values
closing_day_5year = df_tech_five_year['ticker'].loc[df_tech_five_year['date'] == '2018-08-24'].values
full_term_stocks = [stock for stock in opening_day_5year if stock in opening_day_5year and stock in closing_day_5year]
df_tech_five_year = df_tech_five_year[df_tech_five_year['ticker'].isin(full_term_stocks)]
```

```python
df_tech_five_year = df_tech_five_year[['name', 'ticker', 'open', 'close', 'adj_close', 'low', 'high', 'volume']].reset_index(drop=True)
df_tech_five_year.head()
```

```python
df_tech_five_year.shape
```

```python
df_tech_five_year.to_csv('five_year.csv', index=False)
```

```python
# 10 year dataset
```

```python
df_tech_ten_year = df_cleaned_2000.loc[(df_cleaned_2000['date'] >= '2008') & (df_cleaned_2000['sector'] == 'TECHNOLOGY')]
df_tech_ten_year
```

```python
opening_day_10year = df_tech_ten_year['ticker'].loc[df_tech_ten_year['date'] == '2008-01-02'].values
closing_day_10year = df_tech_ten_year['ticker'].loc[df_tech_ten_year['date'] == '2018-08-24'].values
full_term_stocks = [stock for stock in opening_day_10year if stock in opening_day_10year and stock in closing_day_10year]
df_tech_ten_year = df_tech_ten_year[df_tech_ten_year['ticker'].isin(full_term_stocks)]
```

```python
df_tech_ten_year = df_tech_ten_year [['name', 'ticker', 'open', 'close', 'adj_close', 'low', 'high', 'volume']].reset_index(drop=True)
df_tech_ten_year.head()
```

```python
# Add market_signal column with buy or sell based on stock's open/close price per day
df_tech_five_year['market_signal'] = np.where(df_tech_five_year['open'] < df_tech_five_year['close'], 'buy', 'sell')

df_tech_ten_year['market_signal'] = np.where(df_tech_ten_year['open'] < df_tech_ten_year['close'], 'buy', 'sell')
```

```python
# df_tech_five_year.head()
df_tech_five_year.to_csv('five_year.csv', index=False)
```
